```python
# Taking care of cards that are more than ten years. 
# Any customers that have more than 10 years between their start and end date 
# need writing to a separate file, called remove_ccard.json, in the JSON data format. 
```

```python
with open('acw_user_data.csv','r') as customers:
    my_file = csv.reader(customers, delimiter=',')
    next(my_file) # remove the header from the file 
    file_data = []
    for line in my_file:
        #get flat headers
        age                   = line[3]   #['Age']
        distance_commuted     = line[4]   #['Distance Commutted to work']
        employer              = line[5]   #['Employer Company']
        first_name            = line[11]  #['First Name']
        last_name             = line[13]  #['Last Name']
        marital_status        = line[14]  #['Marital Status']
        yearly_pension        = line[15]  #['Yearly Pension']
        retired               = line[16]  #['Retired']
        yearly_salary         = line[17]  #['Yearly Salary']
        sex                   = line[18]  #['Sex']
        vehichle_maker        = line[19] #['Vehicle Make']
        vehicle_model         = line[20] #['Vehicle Model']
        vehicle_year          =line[21]  #['Vehicle Year']
        vehicle_type          = line[22] #['Vehicle Type']
        credit_card_start_date= line[6]  #['Credit Card Start Date']
        credit_card_expiry    = line[7]  #['Credit Card Exipry Date']
        credit_card_number    = line[8]  #['Credit Card Number']
        credit_card_cvv       = line[9]  #['Credit Card CVV']
        credit_card_iban      = line[12] #['Credit Card IBAN']
        address               = line[0]  #['Address Street']
        city                  = line[1]  #['Address City']
        postcode              = line[2]  #['Address Postcode']
        
        # checking if the dependant is empty, assign a default value of 0
        dependant             = line[10]   #['Dependants']
        if dependant == '':
            dependant = 0

            
        # using credit card start date, get the month and year of the card using split function
        ccstart = credit_card_start_date.split('/')
        start_mm = int(ccstart[0]) #credit card start month
        start_yr = int(ccstart[1]) # credit card start year
        
        
        # using credit card exipry date, get the month and year of the card
        ccexpiry = credit_card_expiry.split('/')
        end_mm = int(ccexpiry[0]) #credit card start month
        end_yr = int(ccexpiry[1]) # credit card start year
        
        year_diff = end_yr - start_yr
        
        month_diff = end_mm - start_mm
        
        # The function below gets any card that has more than ten year in-between its start and expiry date
        # if the year difference is greater than 10 year and month difference is less than one month (The card has expired)
        # if the year difference is equal to 10 year and month difference is more than one month (The card is still active)
        # A card issued in June 2013 is still valid as at March 2023
        
        if year_diff >=10 and month_diff <= 0:
            
            #create vehicle nested structure
            vehicle = {'make':vehichle_maker, 'model':vehicle_model, 'year':vehicle_year, 'type':vehicle_type}
            #create card nested structure
            ccard ={'start':credit_card_start_date, 'expiry':credit_card_expiry, 'number':credit_card_number, 'cvv':credit_card_cvv, 'iban':credit_card_iban}
            #create address nested structure
            address= {'address':address, 'city':city, 'postcode':postcode}
            
            #print(f'The card is due {year_diff} and month is {month_diff}')
            
            file_data.append(
            {
                'first_name':first_name, 
                'last_name':last_name,
                'age': age,
                'distance': distance_commuted,
                'employer': employer,
                'marital_status':marital_status, 
                'pension':yearly_pension, 
                'retired':retired, 
                'salary':yearly_salary, 
                'sex':sex ,
                'vehicle':vehicle,
                'credit_card':ccard, 
                'address':address
            })
        
     
    
with open('remove_ccard.json', mode='w') as remove:
    pass



with open('remove_ccard.json', mode='w') as remove:
    json.dump(file_data, remove)
    
```
